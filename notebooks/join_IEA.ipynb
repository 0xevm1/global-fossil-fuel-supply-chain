```python
import glob, os
import pandas as pd
import numpy as np
```

```python
files = glob.glob(os.path.join(os.environ['PYTHONPATH'],'data','iea_balances_public','*'))
```

```python
world_path = files.pop(files.index(os.path.join(os.environ['PYTHONPATH'],'data','iea_balances_public','WORLD.csv')))
```

```python
files.pop(files.index(os.path.join(os.environ['PYTHONPATH'],'data','iea_balances_public','iso_lookup.csv')))
```

```python
files_dict = {f.split('_')[-1][0:2]:f for f in files}
```

```python
df_world = pd.read_csv(world_path).set_index('Unnamed: 0')
```

```python
df_world.sum(axis=1)
```

### Verify that the structure is identical

```python
for kk,vv in files_dict.items():
    #print (kk,vv)
    df = pd.read_csv(vv).set_index('Unnamed: 0').drop(columns=['Total**'])
    #print (kk, list(df_world), list(df))
    print (kk,list(df_world)==list(df), np.prod(df_world.index==df.index))
```

### Find difference between world and all

```python
for kk,vv in files_dict.items():
    print (kk,vv)
    df = pd.read_csv(vv).set_index('Unnamed: 0').drop(columns=['Total**'])
    df_world = df_world - df
```
